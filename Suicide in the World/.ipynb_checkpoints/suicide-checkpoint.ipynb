## Leitura e tratamento dos dados

```python
import pandas as pd
df_suicide=pd.read_csv('master.csv')
```

### Removendo os espaços nos nomes das colunas

```python
df_suicide.columns=df_suicide.columns.str.replace(' ','')
```

### Removendo as colunas que não serão usadas

```python
df_suicide.drop(['HDIforyear','gdp_for_year($)','generation','population','suicides/100kpop'],axis=1,inplace=True)
df_suicide.head()
```

### Verificando se existe algum dado Nulo ou faltando

```python
df_suicide.info()
```

## Qual seria a média de suicídios mundiais de cada ano? E a maior média?

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Agrupando os dados por ano e fazendo a média de cada ano.
df_mean_world=df_suicide.copy().groupby('year').suicides_no.mean() 
df_mean_world=pd.DataFrame(df_mean_world)
df_mean_world.reset_index(inplace=True)
df_mean_world.rename(columns={'suicides_no':'mean_by_year'},inplace=True)
df_mean_world.head()
```

```python
# Criando o gráfico
plt.figure(figsize=(9,6))
plt.bar(df_mean_world['year'],df_mean_world['mean_by_year'],width=0.8)
plt.title('Média de suicídios mundiais de 1985-2016')
plt.ylabel('Quantidade de suicídios')
plt.show()
```

```python
print('Maior média: {}'.format(round(df_mean_world['mean_by_year'].max())))
```

## Em quais anos aproximadamente se obteve o maior número de suicídios?

```python
# Agrupando os dados por anos e fazendo a soma de cada ano.
df_most_year=df_suicide.copy()
df_most_year=df_most_year.groupby(['year']).suicides_no.sum()
df_most_year.head(10)
```

```python
# Criando o gráfico
df_most_year.plot(kind='barh',figsize=(7,8),title='Quantidade de suicídios mundiais',alpha=0.7)
plt.ylabel('')
plt.show()
```

## A diferença de média de suicídios entre o Brasil e os Estados Unidos anuais é muito grande?

```python
# Agrupando os dados por país e anos e fazendo a média.
df_brazil=df_suicide.copy()[['country','suicides_no','year']]
df_brazil=df_brazil.groupby(['country','year']).suicides_no.mean()['Brazil']
```

```python
# Agrupando por país e anos e fazendo a média.
df_us=df_suicide.copy()[['country','suicides_no','year']]
df_us=df_us.groupby(['country','year']).suicides_no.mean()['United States']
```

```python
# Criando o gráfico
plt.figure(figsize=(9,6))
plt.plot(df_brazil,label='Brazil')
plt.plot(df_us,label='US')
plt.title('Média de suicídios do Brazil e Estados Unidos')
plt.ylabel('Média de suicídios')
plt.legend()
plt.show()
```

## Qual a faixa etária mais afetada no mundo desde 1985 à 2016? E no Brasil?

```python
# Agrupando os dados por faixa etária e fazendo a soma de todas ocorrências.
df_age_group=df_suicide.copy().groupby('age').suicides_no.sum()
df_age_group=pd.DataFrame(df_age_group)
df_age_group.reset_index(inplace=True)
df_age_group.rename(columns={'suicides_no':'total'},inplace=True)
df_age_group
```

```python
# Função que troca a palavra years por anos
def change(labels):
    for i in range(len(labels)):
        labels[i]=labels[i].replace('years','anos')
    return labels

labels_names=df_age_group['age'].tolist()
labels_names=change(labels_names)
```

```python
# Criando gráfico
plt.figure(figsize=(9,6))
plt.pie(df_age_group['total'],labels=labels_names,shadow=True,autopct='%1.1f%%') 
plt.title('Faixas etárias mais afetadas no mundo desde 1985-2016')
plt.axis('equal')
plt.show()
```

```python
# Agrupando os daods por país e faixa etária e fazendo a soma de todas ocorrências.
df_brazil_group=df_suicide.copy().groupby(['country','age']).suicides_no.sum()['Brazil']
df_brazil_group=pd.DataFrame(df_brazil_group)
df_brazil_group.reset_index(inplace=True)
df_brazil_group.rename(columns={'suicides_no':'total'},inplace=True)
```

```python
# Criando o gráfico
plt.figure(figsize=(9,6))
labels_names=change(df_age_group['age'].tolist())
plt.pie(df_brazil_group['total'],labels=labels_names,shadow=True,autopct='%1.1f%%')
plt.title('Faixa etárias mais afetadas no Brasil desde 1985-2016')
plt.axis('equal')
plt.show()
```

## Quantas pessoas de ambos os sexos no Brasil cometeram tal ato se compararmos 1985 à 2015?

```python
# Coletando todas as ocorrências onde temos a coluna year igual aos anos de 1985 e 2015, agrupando os dados por ano e sexo e somando as ocorrências de acordo com cada ano e sexo.
df_sex=df_suicide.copy().query('year==1985 or year==2015').groupby(['year','sex']).suicides_no.sum()
df_sex
```

```python
# Função que coloca automaticamente o valor de cada barra no topo dela
def val_bar(xlocs,y):
    for i, v in enumerate(y):
        plt.text(xlocs[i] - 0.30, v + 2000, str(v),fontweight='bold')
```

```python
# Criando o gráfico
plt.figure(figsize=(9,6))
sex=['Mulher','Homem']
xpos=np.arange(len(sex))
plt.xticks(xpos+0.2,sex)
plt.bar(xpos,df_sex[1985],width=0.4,label='1985')
plt.bar(xpos+0.4,df_sex[2015],width=0.4,label='2015')
val_bar(xpos+0.2,df_sex[1985])
val_bar(xpos+0.6,df_sex[2015])
plt.title('Suicídios de ambos os sexos em 1985 e 2015')
plt.ylabel('Quantidade de suicídios')
plt.legend()
plt.show()
```
