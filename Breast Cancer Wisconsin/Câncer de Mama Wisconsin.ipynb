# Câncer de Mama Wisconsin

### Objetivo:
Utilizar Machine Learning para diagnóstico de câncer de mama com base nas características do câncer.

### Dataset:

O dataset foi criado apartir de imagens da massa da mama. A qual descreve características do nucleo da célula presente na imagem.

### Informações sobre os atributos do dataset:
1- ID<br/>
2- Diagnóstico (M = Maligno, B = Benigno) <br/>
3- Valores que foram extraídos das imagens:<br/>

* raio (média das distancias do centro até os pontos do perímetro) 
* textura (desvio padrão da escala cinza) 
* perímetro 
* área 
* suavidade (variação local no comprimento do raio) 
* compacidade ($\frac{perímetro^2}{área - 1.0}$) 
* concavidade (gravidade das porções côncavas do contorno) 
* pontos de concavidade (número de porções côncavas do contorno) 
* simetria 
* dimensão fractal ("aproximação costa" - 1) 

#### Dataset utilizado: [Breast Cancer Wisconsin (Diagnostic) Data Set](https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%29)

## Análise das características

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
df.head()
```

Checando se há valores nulos no dataset

```python
df.info()
```

### Visualizando os dados

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Verificando a proporção de cada tipo de câncer
df_prop=df.copy()['diagnosis']
df_prop=pd.get_dummies(df_prop)
df_prop.rename(columns={'B':'Benigno','M':'Maligno'},inplace=True)
df_prop=df_prop.sum()
df_prop
```

```python
# Função que coloca automaticamente o valor de cada barra no topo dela
def val_bar(xlocs,y):
    for i, v in enumerate(y):
        plt.text(xlocs[i] - 0.30, v + 2000, str(v),fontweight='bold')
```

```python
#Criando o gráfico
df_prop.plot(kind='bar',figsize=(9,6),title='Proporção de câncer malignos e benignos',rot=0)
plt.ylabel('Quantidade de ocorrências')
plt.show()
```

Podemos ver que o câncer benigno possui uma proporção maior do que o maligno.

```python
# Criando o mapa de correlação
# Acabei deixando apenas um número depois da vírgula
plt.figure(figsize=(16,10))
sns.heatmap(df.corr(),annot=True,cmap='viridis',fmt = '.1f')
plt.show()
```

Assim podemos verificar quantas aproximadamente características iremos utilizar.

## Pré-Processamento

```python
# Reservei 20% para os testes e 80% para o treinamento
from sklearn.model_selection import train_test_split

X=df.iloc[:,1:]
y=df['diagnosis']
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=0)
print('Training Features Shape: {}'.format(X_train.shape))
print('Training Target Shape: {}'.format(y_train.shape))
print('Test Features Shape: {}'.format(X_test.shape))
print('Test Target Shape: {}'.format(y_test.shape))
```

## PCA (Principal Component Analysis)

```python
from sklearn.decomposition import PCA
# Delimitando quantos componentes terá
pca=PCA(n_components=10)
# Treinando e redimensionando os features
pca.fit_transform(X_train)
# Redimensionando os features
pca.transform(X_test)
```

## SVC (Support Vector Classification)

```python
# Criando uma grade de busca para os melhores parametros que podem ser utilizados com SVC
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

parameters={'kernel':('linear','rbf'),'C':[1, 10, 100, 1000],'gamma':[0.0001, 0.1, 1, 10, 1000]}
svc=GridSearchCV(SVC(),parameters,cv=5)
svc.fit(X_train,y_train)
best_svc=svc.best_estimator_
```

```python
best_svc
```

```python
print('Acurácia no treinamento: {}'.format(best_svc.score(X_test,y_test)))
print('Acurácia no teste: {}'.format(best_svc.score(X_train,y_train)))
```

```python
# Fazendo previsões e criando matriz de confusão para exibição dos dados obtidos do modelo
from sklearn.metrics import confusion_matrix

y_pred=best_svc.predict(X_test)
confusion=confusion_matrix(y_test,y_pred)
confusion
```

```python
types=['Benigno','Maligno']
sns.heatmap(confusion, annot=True, cmap=plt.cm.Blues, xticklabels=types, yticklabels=types)
plt.xlabel('Valor previsto')
plt.ylabel('Valor verdadeiro')
plt.show()
```

## Naive Bayes

```python
# Utilizando Gaussian Naive Bayes para criação de modelo preditivo
from sklearn.naive_bayes import GaussianNB

nb=GaussianNB()
nb.fit(X_train,y_train)
```

```python
print('Acurácia no treinamento: {}'.format(nb.score(X_test,y_test)))
print('Acurácia no teste: {}'.format(nb.score(X_train,y_train)))
```

Como podemos ver a acurácia utilizando Naive Bayes não foi tão boa quanto com Support Vector Classification

## Random Forest

```python
# Utilizando uma grade de busca para os melhores parametros que foram passados que devolvam uma acurácia alta utilizando
# Random Forest para classificação
from sklearn.ensemble import RandomForestClassifier

parameters={'n_estimators':[10,100],'criterion':['gini','entropy']}
rf=GridSearchCV(RandomForestClassifier(),parameters,cv=4)
rf.fit(X_train,y_train)
best_rf=rf.best_estimator_
```

```python
best_rf
```

```python
print('Acurácia no treinamento: {}'.format(best_rf.score(X_test,y_test)))
print('Acurácia no teste: {}'.format(best_rf.score(X_train,y_train)))
```

Aparentemente obtemos uma melhora significativa na acurácia no teste e no treinamento

```python
# Fazendo previsões e utilizando matriz de confusão para exibição dos dados obtidos do modelo
y_pred=best_rf.predict(X_test)
confusion=confusion_matrix(y_test,y_pred)
confusion
```

```python
types=['Benigno','Maligno']
sns.heatmap(confusion, annot=True, cmap=plt.cm.Blues, xticklabels=types, yticklabels=types)
plt.xlabel('Valor previsto')
plt.ylabel('Valor verdadeiro')
plt.show()
```

## Conclusão:
O que se teve maior perfomance dentre SVC, Naive Bayes e Random Forest foi Random Forest, claro que existem melhorias que podem ser feitas e se possível entre em contato para aplica-las.
